```python
# Lowercase characters
df['text']=df['text'].str.lower()
df.head(10)
```

```python
# Remove floating point numbers and numerical values from the 'text' column
def remove_numbers(text):
    # Using regex to remove any digits or floating point numbers
    cleaned_text = re.sub(r'\d+(\.\d+)?', '', text)
    return cleaned_text

# Applying the remove_numbers function to the 'text' column
df['text'] = df['text'].astype(str)
df['text'] = df['text'].apply(remove_numbers)

df.head(10)
```

```python
# Remove emojis from the 'text' column
def remove_emojis(text):
    # Use regex to remove emojis
    cleaned_text = re.sub(r'[^\w\s,]', '', text)
    return cleaned_text

# Apply the remove_emojis function to the 'text' column
df['text'] = df['text'].astype(str)
df['text'] = df['text'].apply(remove_emojis)

# Save the cleaned DataFrame to a new CSV file
df.head(10)
```
